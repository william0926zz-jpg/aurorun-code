```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
client = OpenAI(
    base_url="https://api.chatanywhere.tech"
)
```

```python
## SECTION 2: Inference -> Predict next token
import tiktoken
enc = tiktoken.get_encoding("cl100k_base")

prompt = "The patient was admitted because of severe abdominal"

generated_text = prompt
for _ in range(10):  # Predict 10 tokens one by one
    response = client.completions.create(
        model="gpt-4o-mini",
        prompt=generated_text,
        max_tokens=1,   # predict one token
        temperature=0
    )

    next_token = response.choices[0].text
    generated_text += next_token

    token_id = enc.encode(next_token)
    print(f"Predicted token: '{next_token}' (ID: {token_id})")
```
